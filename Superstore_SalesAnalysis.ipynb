# EXPLORATORY DATA ANALYSIS

<h3> Q1. what is the total sales trend?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
# Getting month year from order_date
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# grouping month_year by sales
df_temp = df.groupby('month_year')['sales'].sum().reset_index()
```

```python
# Setting the figure size
plt.figure(figsize=(15,6))
plt.plot(df_temp['month_year'], df_temp['sales'], color = 'red')
plt.xticks(rotation = 'vertical',size = 8)
plt.show()
```

<h3> Q2. which are the top 10 products by sales?

```python
# Grouping products by sales
product_sales = pd.DataFrame(df.groupby('product_name')['sales'].sum())
```

```python
# Sorting the dataframe in descending order
product_sales = product_sales.sort_values('sales',ascending = False)
```

```python
# Top 10 products by sales
product_sales[:10]
```

<h3>Q3. which are the most selling products?

```python
# Grouping products by Quantity
most_sales_prod = pd.DataFrame(df.groupby('product_name')['quantity'].sum())
```

```python
# Sorting the dataframe in descending order
most_sales_prod = most_sales_prod.sort_values('quantity', ascending=False)
```

```python
# Most selling products
most_sales_prod[:10]
```

<h3>Q4. What is the most preferred ship mode?

```python
# Setting the figure size
plt.figure(figsize=(8, 5))

# Create a colorful countplot with hue and disable the legend
sns.countplot(x='ship_mode', data=df, hue='ship_mode', palette='Set2', legend=False)

# Adding title and labels
plt.title('Preferred Shipping Mode', fontsize=16)
plt.xlabel('Shipping Mode', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Display the figure
plt.show()
```

### Explanation:
- hue='ship_mode': Assigns the color to the categories of the ship_mode column.
- legend=False: Since hue creates a legend by default, we turn it off to keep the plot simple.

<h3>Q5. which are the most profitable category and sub-category?

```python
# Group by 'category' and 'sub_category', and sum the 'profit'
cat_subcat = pd.DataFrame(df.groupby(['category', 'sub_category'])['profit'].sum().reset_index())

# Sort by 'category' and 'profit' in descending order
cat_subcat = cat_subcat.sort_values(['category', 'profit'], ascending=False)
cat_subcat
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(x = 'sub_category', y = 'profit', hue = 'category', data=cat_subcat)
plt.xticks(rotation = 'vertical',size = 8)

# Adding title and labels
plt.title('Profitable Category and Sub-category', fontsize=16)
plt.xlabel('Sub-category', fontsize=12)
plt.ylabel('Profit', fontsize=12)

plt.show()
```

<h3>Q6. Which are the Top 10 Countries?

```python
# Group by country and sum the sales, then sort in descending order and take the top 10
top_countries = df.groupby('country')['sales'].sum().sort_values(ascending=False).head(10)

# Convert the sales to a more readable format
top_countries_formatted = top_countries.apply(lambda x: f"${x:,.2f}")

# Display the result
countries_sales = pd.DataFrame(top_countries_formatted)
countries_sales
```

<h3>Q7. Which are the Top 10 Costumers?

```python
# Group by customer name and sum the sales, then sort in descending order and take the top 10 customers
top_customers = df.groupby('customer_name')['sales'].sum().sort_values(ascending=False).head(10)

# Convert the sales to a more readable format
top_customers_formatted = pd.DataFrame(top_customers.apply(lambda x: f"${x:,.2f}"))

top_customers_formatted
```
